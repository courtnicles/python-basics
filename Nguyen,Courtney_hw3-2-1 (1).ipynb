```python
import json
import urllib
```

### Open URLs for each day in August  

```python
link = []
for i in range(1,32):
    day_link = "https://www.metaweather.com/api/location/2487956/2019/08/" + str(i)
    link.append(day_link)
```

```python
print(link)
```

### Load each dataset

```python
json_list = []
for i in link:
    response = urllib.request.urlopen(i)
    weather_read = response.read()
    data_str = weather_read.decode("utf-8") 
    json_list.append(json.loads(data_str))
```

```python
print(json_list[0][1]['created'])
```

```python
weather_by_day = {}
for i in json_list:
    for x in i:
        if x['applicable_date'] in x['created']:
            weather_by_day[x['created'] + ", " + x['applicable_date']] = weather_by_day.get(x['created'], 0) + int(x['the_temp'])
```

```python
for k,v in weather_by_day.items():
    print(k,v)
```

### Set up for average weather for each day

```python
weather_by_day_av = {}
av_counter = []
for i in json_list:
    for x in i:
        if x['applicable_date'] in x['created']:
            weather_by_day_av[x['applicable_date']] = weather_by_day_av.get(x['applicable_date'], 0) + int(x['the_temp'])
            count_of_keys = len(weather_by_day_av)
            av_counter.append(count_of_keys)
total = len(weather_by_day_av)
list_of_count = []
for i in range(1,(total+1)):
    av = av_counter.count(i)
    list_of_count.append(av)
```

```python
i = 0
final_dict = {}
for k,v in weather_by_day_av.items():
    final_dict[k] = v/list_of_count[i]
    print(str(k) + ", " + str(v/list_of_count[i]))
    i += 1
```

### Save results to file

```python
with open('aver_temp.json', 'w') as f:
    json.dump(final_dict, f)
```
